```python
import os
import pandas as pd

# Base folder where your files live
BASE_FOLDER = r"C:\Users\Julian.Diaz\OneDrive - XENITH CONSULTING PTY LTD\Documents\05_Geodatabases\05_Excel"

print("Looking in folder:")
print(BASE_FOLDER)
print("\nFiles found in this folder:")
for f in os.listdir(BASE_FOLDER):
    print(" -", f)

# Filenames (change if needed to match exactly)
west_name = "NWQLD_West_Drilling-Collars.csv"
east_name = "NWQLD_East_Drilling-Collars.csv"

west_file = os.path.join(BASE_FOLDER, west_name)
east_file = os.path.join(BASE_FOLDER, east_name)

# Check that the files actually exist
for path in [west_file, east_file]:
    if not os.path.exists(path):
        raise FileNotFoundError(
            f"\nFile not found: {path}\n"
            f"Check the exact name & extension in the list printed above."
        )

# Helper to read CSV with a couple of likely encodings
def read_csv_safely(path):
    for enc in ("utf-8", "cp1252", "latin1"):
        try:
            print(f"\nReading {os.path.basename(path)} with encoding='{enc}'")
            return pd.read_csv(path, encoding=enc)
        except UnicodeDecodeError as e:
            print(f"  Failed with encoding '{enc}': {e}")
    # If all encodings fail, re-raise the last error
    # (but latin1 should basically always work)
    return pd.read_csv(path, encoding="latin1")

# Read the CSVs
df_west = read_csv_safely(west_file)
df_east = read_csv_safely(east_file)

# Concatenate (stack rows)
merged_df = pd.concat([df_west, df_east], ignore_index=True)
print(f"\nMerged rows: {len(merged_df)}")

# Save to new CSV in the same folder
output_file = os.path.join(BASE_FOLDER, "NWQLD_EW-Merge_Drilling-Collars.csv")
merged_df.to_csv(output_file, index=False)

print(f"\nMerged file saved to: {output_file}")
```
